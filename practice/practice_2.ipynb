### Requirements 

- Make a binary classification dataset with Scikit-Learn's make_moon() function.
- Turn the data into Pytorch tensors. Split the data into training and test sets using train_test_split with 80% training and 20% testing


```python
import torch
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
RANDOM_SEED = 42

X_moon, y_moon = make_moons(n_samples= 1000, 
                            noise= 0.1, 
                            random_state= RANDOM_SEED)

# Data are suspiciously numpy arrays, i have to convert them to Tensors
X_moon = torch.from_numpy(X_moon).type(torch.LongTensor)
y_moon = torch.from_numpy(y_moon).type(torch.LongTensor)

# Split data into 80% training and 20% testing as required
X_moon_test, y_moon_test, X_moon_train, y_moon_train = train_test_split(X_moon,
                                            y_moon,
                                            test_size= 0.2,
                                            random_state= RANDOM_SEED)

# isualising data
plt.figure(figsize=(10, 7))
#plt.scatter(X_moon[:, 0], X_moon[:, 1], c=y_moon, cmap=plt.cm.RdYlBu)
plt.scatter(X_moon[:, 0], X_moon[:, 1], c=y_moon, cmap="coolwarm", edgecolors="k")

# 3. Add labels and show
plt.title("Make Moons Scatter Plot")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
