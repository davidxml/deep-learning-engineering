### Import Neccessary Libraries

```python
import torch
from torch import nn
```

### Data generation


```python
# Create *known* parameters    
weight = 0.7
bias = 0.3

# Create
start = 0
end = 1 
step = 0.02
X = torch.arange(start, end, step).unsqueeze(dim=1)
y =  weight  * X + bias

X[:10], y[:10] 
```

### Data splitting 

```python
split = int(0.8 * len(X))

X_train, y_train = X[:split], y[:split]
X_test, y_test = X[split:], y[split:]

```

### Model Building


```python
class PracticeModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer_1 = nn.Linear(in_features= 1, out_features=10)
        self.layer_2 = nn.Linear(in_features= 10, out_features=10)
        self.layer_3 = nn.Linear(in_features= 10, out_features= 1)

    def forward(self, x):
        return self.layer_3(self.layer_2(self.layer_1(x)))
    
# instantiate Model
model_1 =  PracticeModel()
model_1
```

### Model Training

```python
optimizer =  torch.optim.SGD(params= model_1.parameters(),
                             lr = 0.1)

loss_fn  = nn.BCEWithLogitsLoss()

def accuracy_fn(y_true, y_preds):
    correct = torch.eq(y_true, y_preds).sum().item()
    acc = (correct/len(y_preds)) * 100
    return acc
```

```python
epochs = 100

for epoch in range(epochs):
    model_1.train()

    y_logits = model_1(X_train)
    y_prob = torch.sigmoid(y_logits)
    acc = accuracy_fn(y_true = y_train,
                      y_preds = y_prob)

    loss = loss_fn(y_logits, 
                   y_train) 
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Testing
    model_1.eval()

    with torch.inference_mode():
        test_logits = model_1(X_test)
        test_prop = torch.sigmoid(test_logits)

        test_loss = loss_fn(test_prop, y_test)
        test_accuracy = accuracy_fn(y_true = y_test,
                                    y_preds = test_prop)
    
        
        if epoch % 10 == 0:
            print(f"Epoch is {epoch} | Training loss = {loss:.2f} | Training accuracy = {acc} | Test accuracy = {test_accuracy}")
```
