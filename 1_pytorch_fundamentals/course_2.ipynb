 ### Reproducibility In PyTorch (trying to make random out of random)

How a neural network is:

` Start with random numbers --> tensor operations --> Update random numbers to make them represent the data best --> Repeat...`

To reduce the randomness in neural networks and PyTorch comes the concept of **random seeds**. For PyTorch it is `torch.manual_seed(Random_Seed)`

Essentially, what the random seed does is flavour the randomness

```python
import torch 

rand_tensorA =  torch.rand (3, 4)
rand_tensorB =  torch.rand (3, 4)

print(rand_tensorA)
print(rand_tensorB)

print(rand_tensorA == rand_tensorB) 
```

```python
# Reproducible random tensors
import torch 

# Set random seed
RANDOM_SEED = 42        # could be any integer ...42 is only a standard practice 
torch.manual_seed(RANDOM_SEED) 
rand_tensorC = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED) 
rand_tensorD = torch.rand(3, 4)

print(rand_tensorC)
print(rand_tensorD)

print(rand_tensorC == rand_tensorD)
```

### Putting Tensors and models on the GPU

The reason for putting on a GPU is because they result in faster computations

```python
# Create a Tensor, default is created on the cpu

tensor = torch.tensor([1, 2, 3])

print(tensor, tensor.device)
```

```python
# Move tensors to Gpu (if available)

tensor_on_gpu = tensor.to(device)

tensor_on_gpu
```

### Moving tensors back to the cpu  

```python
tensor_on_gpu.numpy() 

""" This worked on this laptop because the cause tensor_on_gpu was
    initialised to tensor.to(device) and the device was set 
    to cpu in the absence of a GPU."""

```
